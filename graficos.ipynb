```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
from datetime import datetime
```

### **<font color='purple'>Leitura da base de dados</font>**

```python
df = pd.read_csv('dados/VC_ST_dia.csv', sep=';', index_col='data')
df.head()
```

### **<font color='purple'>Transformações na coluna de data</font>**

```python
df.index = pd.to_datetime(df.index, format='%d/%m/%Y')
df.index
```

### **<font color='purple'>Visualização da série temporal</font>**

```python
df.plot(figsize=(8,5))
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)

df.plot(ax=ax, style='k-')

outliers = [
    (datetime(2022, 7, 11), 'Absolutely no idea'),
    (datetime(2022, 10, 15), 'Acessos post AJ')
]

for date, label in outliers:
    ax.annotate(label, xy=(date, df.asof(date) + 300),
                xytext = (date, df.asof(date) + 825),
                arrowprops=dict(facecolor='blue', headwidth=4, width=2, 
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

    ax.set_xlim(['1/1/2022', '31/12/2022'])
    ax.set_ylim([10, 1500])
    ax.set_title('Dias com possíveis outliers')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.random(50).cumsum(),'g--')
ax1.hist(np.random.random(100), bins = 20, color = 'g', alpha = 0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30), color = "orangered")
ax3.scatter(np.arange(50), np.arange(50) + 3 * np.random.randn(50))
```
